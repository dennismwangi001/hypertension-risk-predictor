# Presentation: Choosing the Best AI Model for the Clinic

## **The Bottom Line**
If we are deploying an AI tool to screen for hypertension in a resource-limited Kenyan clinic, the clear winner is **XGBoost**. 

Here is a simple breakdown of why XGBoost beats the other models, based on four key factors:

---

### **1. Accuracy: Catching the Disease (Predictive Performance)**
*In healthcare, the biggest risk is missing a sick patient. In data science, we call this "Recall."*

* **XGBoost (The Winner):** It correctly identifies about **92%** of patients who actually have hypertension. 
* **Random Forest:** It does well, catching about **90%** of patients, but XGBoost is slightly more accurate and makes fewer mistakes overall.
* **KNN:** This model catches almost everyone (**98%**), but it creates too many "false alarms" (false positives). In a resource-limited clinic, we don't have the staff or money to constantly re-test healthy people who were falsely flagged.
* **Logistic Regression:** This model only catches about **68%** of patients. This means it would **miss nearly 1 in 3 sick patients**, which is simply too risky for a medical screening tool.
* **SVM & Decision Tree:** These sit in the middle (catching around 81-87%), but they don't beat XGBoost's overall balance of accuracy.

### **2. Understanding the Results (Clinical Interpretability)**
*Doctors need to trust the AI. They need to know **why** it flagged a patient.*

* **Logistic Regression** is very easy to understand, but as we saw, it misses too many patients. 
* **XGBoost** is more complex. It works by combining hundreds of small decision trees, which makes it a "black box." 
* **The Solution:** We don't have to rely on a black box. We can use a simple visual tool (called SHAP) that plugs into XGBoost. It generates a simple receipt for the doctor that says: *"This patient was flagged as high risk primarily because of their Age and BMI."* This gives the doctor the accuracy of XGBoost with the transparency they need to trust it.

### **3. Computer Power and Speed (Computational Requirements)**
*Does this require an expensive supercomputer?*

* **No.** Even though XGBoost is a powerful, advanced model, it is specifically designed to be incredibly fast and lightweight. 
* It does not require massive servers or cloud computing. It can easily run on a standard, low-cost clinic laptop, or even on a basic smartphone or tablet used by Community Health Promoters (CHPs) doing door-to-door screenings in rural areas.

### **4. Real-World Clinic Use (Implementation & Maintenance)**
*Is it easy to put into the clinic's daily workflow?*

* **Easy to Integrate:** XGBoost models are very small files. They can be easily plugged into the electronic medical record (EMR) systems already used in many Kenyan clinics, like OpenMRS or KenyaEMR. 
* **Easy to Update:** As the clinic sees more patients over the next few years, the clinic's IT team can easily feed that new data back into the model to keep it updated and accurate for the local population.

---

## **Summary / Closing Statement**
"While simpler models like Logistic Regression are easier to understand, they miss too many patients. More complex models like KNN create too many false alarms. **XGBoost** gives us the best of both worlds: it catches the vast majority of sick patients, runs perfectly on basic clinic hardware, and can be easily explained to doctors using visual tools. Therefore, XGBoost is the safest and most practical choice for deployment."

Part 1: The Data Science Report
Executive Summary
This report outlines the development of a predictive machine learning model designed to identify patients at high risk of hypertension. By analyzing patient health records, we built a system that accurately flags at-risk individuals, allowing for early medical intervention. The final solution utilizes advanced "ensemble" algorithms (XGBoost and Random Forest) which demonstrated superior ability to identify true positive cases without generating excessive false alarms.
1. Understanding the Data & The Problem
We started with a dataset of 4,999 patient records containing 17 different health and demographic variables.
The Challenge: The data was heavily "imbalanced." 92.2% of patients did not have hypertension, while only 7.8% did.
Why this matters: If a model simply guessed "No Hypertension" for every single patient, it would be 92% accurate, but completely useless in a hospital because it would miss every single sick patient.
2. Data Preparation: Cleaning the Canvas
Before building the model, we had to clean the data. Here is why we made specific choices:
Dropping Specific Columns (SBP, DBP, SBP_ge120, ID): We intentionally removed Systolic and Diastolic Blood Pressure readings. Why? Because our target variable (Hypertension) is literally defined by these numbers. Including them is like giving a student the answer key before a test (a concept called Data Leakage). We also dropped ID as it is just a random number with no medical value.
Handling Missing Values: Some patient records had missing lab results. Instead of deleting those patients (which would shrink our dataset), we used Imputation. We filled numerical gaps with the "median" (middle value) and categorical gaps with the "most frequent" value.
Feature Scaling: We standardized numerical data (like Age and BMI) so they are on the same mathematical scale. Why? Some algorithms get confused if one feature ranges from 0-100 (Age) and another ranges from 0-1 (Gender).
3. Solving the Imbalance Problem (SMOTE)
To stop the model from just guessing the majority class, we used a technique called SMOTE (Synthetic Minority Over-sampling Technique).
What it does: It creates realistic, synthetic "fake" data points for the minority class (hypertension patients) to balance the scales.
Crucial Rule: We only applied SMOTE to the Training Data, never the Testing Data. Why? If we apply it to the whole dataset before splitting it, information from the test set "leaks" into the training set, giving us falsely high performance scores.
4. Choosing the Right "Scorecard" (Evaluation Metrics)
We did not use "Accuracy" as our main metric because of the data imbalance mentioned above. Instead, we used:
Recall (Sensitivity): Out of all the actual hypertension patients, how many did the model correctly find? (Crucial for healthcare; we don't want to miss sick people).
Precision: Out of all the people the model said had hypertension, how many actually did? (Important so we don't waste medical resources on false alarms).
ROC-AUC: A score from 0 to 1 that measures the model's overall ability to distinguish between the two classes.
5. Model Selection: Why We Chose the Winners
We tested 7 different algorithms, ranging from simple (Logistic Regression) to complex (XGBoost).
Why we didn't choose the simpler models: Logistic Regression and Naive Bayes were too simplistic to capture the complex, non-linear relationships in human biology. Decision Trees were prone to "overfitting" (memorizing the training data but failing on new data).
Why XGBoost and Random Forest won: These are "Ensemble" methods. Instead of relying on one decision tree, they build hundreds of trees and let them "vote" on the outcome. This makes them incredibly robust, highly accurate, and excellent at handling imbalanced data. XGBoost achieved the highest ROC-AUC (0.956), making it our top performer.
6. Ensuring Reliability (Cross-Validation)
To prove our model wasn't just "lucky" with one specific split of data, we used 5-Fold Cross-Validation.
Analogy: Instead of taking one practice test, the model takes 5 different practice tests covering different parts of the curriculum. XGBoost and Random Forest consistently scored high across all 5 tests, proving they are reliable for real-world deployment.
7. Conclusion
By carefully preventing data leakage, correcting class imbalance, and utilizing advanced ensemble algorithms, we have developed a highly reliable predictive tool. This model will allow healthcare providers to proactively identify patients at risk of hypertension, shifting the focus from reactive treatment to proactive prevention.
